#### Testing Club/ Player Accuracy

Next step is to manually verify the entries in excel and re open to calculate accuracy

```python
df = pd.read_csv("C:/Users/Titoran/Documents/Testing/Sentiment Accuracy.csv", index_col = [0])
```

```python
df
```

```python
n = 0 
count_clubs = 0
count_players = 0
count_sentiments = 0

for i in range(len(df)) : 
    if df.iloc[i,1] == df.iloc[i,2] : 
        count_clubs += 1
    
    if df.iloc[i,3] == df.iloc[i,4] : 
        count_players += 1 
        
    if df.iloc[i,5] == df.iloc[i,6] : 
        count_sentiments += 1
```

```python
accuracy_clubs = round(count_clubs/len(df) * 100,2)
```

```python
accuracy_players = round(count_players/len(df)*100,2)
```

```python
accuracy_sentiment = round(count_sentiments/len(df)*100,2)
```

```python
print("club accuracy : " + str(accuracy_clubs))
print("players accuracy : " + str(accuracy_players))
print("sentiment accuracy : " + str(accuracy_sentiment))
```
